```python
# Import pandas and geopandas
import pandas as pd
import geopandas as gpd
```

```python
# use wget to download the GLODAPv2.2023 master file zip archive "GLODAPv2.2023_Merged_Master_File.csv.zip"
!wget https://glodap.info/glodap_files/v2.2023/GLODAPv2.2023_Merged_Master_File.csv.zip
```

```python
# Load Glodap data into a pandas dataframe 
# hinting at datatypes in the glodap csv file fields

df = pd.read_csv('GLODAPv2.2023_Merged_Master_File.csv.zip',dtype={'G2expocode' : str, 'G2Cruise' : int, 'G2region' :int , 'G2cast' :int ,'G2year': int,'G2month' : int,'G2day' : int,	'G2hour': int, 'G2minute' : int, 'G2bottle' : int, 'G2doi' : str})
```

```python
# examine the data frame
df.head(10)
```

```python
df.info()
```

```python
# Remove the first 2 "G2" characters from column names
df.columns = df.columns.str[2:]
# examine the fields and sample data
print(df.head())
```

```python
# Combine year, month, day, hour, and minute columns into a new datetime column
df['recordedtime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute']])
```

```python
from shapely.geometry import Point

# Create a GeoPandas GeoDataFrame with Point geometries
gdf = gpd.GeoDataFrame(df, 
                        geometry=gpd.points_from_xy(df['longitude'], df['latitude']),
                        crs='EPSG:4326')  # Set the coordinate reference system (CRS)
```

```python
# Install the pyarrow library dependency for geoparquet export from geopandas
%pip install pyarrow
```

```python
# export the geopandas dataframe as geoparquet
gdf.to_parquet('glodap_v2_2023_merged_master.parquet')
```
