```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Jupyter").getOrCreate()
spark.conf.set("spark.sql.legacy.parquet.nanosAsLong", "true")

spark
```

```python
%%sql

CREATE DATABASE IF NOT EXISTS glodap
```

```python
%%sql

DROP TABLE IF EXISTS glodap.glodap_v2_2023
```

```python
df = spark.read.parquet("/home/iceberg/data/glodap_v2_2023_merged_master.parquet")
df.write.saveAsTable("glodap.glodap_v2_2023")
```

```python
%%sql

DESCRIBE EXTENDED glodap.glodap_v2_2023
```

```python
%%sql

SELECT COUNT(*) as cnt
FROM glodap.glodap_v2_2023
```

```python
%%sql

SELECT doi,latitude,longitude,depth,MAX(temperature) FROM glodap.glodap_v2_2023 GROUP BY doi,latitude,longitude,depth ORDER BY MAX(temperature) DESC;
```

```python
%%sql

SELECT doi, MAX(temperature) FROM glodap.glodap_v2_2023 GROUP BY doi ORDER BY MAX(temperature) DESC;
```
